```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
class my_multiple_liner_reg:
    def __init__(self):
        self.coef_=None
        self.offset=None

    def fit(self,X_train,y_train):
        X_train=np.insert(X_train,0,1,axis=1)
        beta=np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.offset=beta[0]
        self.coef_=beta[1:]
        return self.offset,self.coef_

    def predict(self,X_test):
         y_pred = np.dot(X_test,self.coef_) + self.offset


         return y_pred

        
```

```python
X1=X_train
```

```python
X1.shape
```

```python
sr=my_multiple_liner_reg()
sr.fit(X1,y_train)
```

```python
Y1=sr.predict(X_test)
r2_score(y_test,Y1)
```
